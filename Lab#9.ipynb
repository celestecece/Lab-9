```python
importing libraries
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv (r'./example_export.txt', sep = '\t') 
df
```

```python
fig, ax = plt.subplots()
df.plot.line(ax = ax, y = 'EMG', x = 'Time')
ax.set_xlabel("Time (Seconds)")
ax.set_ylabel("EMG (Volts)")
```

```python
window_size = 3  # Window size in seconds (how large of a window you want to take the avg max of)
step_size = 10  # Step size in seconds (how often you want to check the window size)
```

```python
max_values = []
time_intervals = []


for i in range(0, len(df), int(step_size / 0.001)):
    start_idx = i
    end_idx = start_idx + int(window_size / 0.001)
    
    if end_idx < len(df):
        window_data = df.iloc[start_idx:end_idx]
        max_value = window_data['EMG'].max()
        max_values.append(max_value)
        
       
        time_interval = window_data['Time'].mean()
        time_intervals.append(time_interval)


result_df = pd.DataFrame({'Time': time_intervals, 'MaxEMG': max_values})


result_df['MaxEMG_percent'] = (result_df['MaxEMG'] / result_df['MaxEMG'].max()) * 100

# Display the result
print(result_df)
```

```python
threshold = 70
```

```python
reached_threshold = result_df[result_df['MaxEMG_percent'] <= threshold]
time_at_threshold = reached_threshold['Time'].iloc[0]
print(f"Time at ~{threshold} % of max: {time_at_threshold} seconds")
```

```python
#Discussion:
The phases of a typical skeltal muscle twitch involves observing and analyzing the changes in a muscle acitivity during 
a contraction. Some of the key phases are 1.)Latent Phase: Which is hard to tell, because there is no visible muscle activity occurs during this time.
The electrical events are taking place in preparation for contraction.
The contraction phase is the period of time when the muscle is actively contracting 
in response to the stimulus. The tension builds as cross-bridges form between actin and myosin filaments, 
leading to the shortening of sarcomeres and muscle contraction.
The relaxation phase that follows the contraction phase, the muscle returns to its original length as cross-bridges detach,
and tension decreases.Calcium ions are actively pumped back into the sarcoplasmic reticulum, that allows for muscle relaxation.
```
